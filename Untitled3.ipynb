```python
# ======================================================================
# Statistics 101 Mini-Project – Colab Ready
# Topic: Does more study time lead to better exam scores?
# Covers most modules: descriptive stats, visualization, correlation,
# regression, skewness/kurtosis, hypothesis testing + mini-project style
# Current date reference: February 2026
# ======================================================================

# ── 1. Libraries (all pre-installed in Colab) ────────────────────────
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm

sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
%matplotlib inline

print("Setup complete ✓\n")

# ── 2. Create realistic sample dataset (replace with your real data) ──
np.random.seed(42)

n_students = 50

study_hours = np.random.normal(11, 5, n_students).clip(1, 28).round(1)

# Simulated relationship: more hours → better score + noise
exam_score = 52 + 2.3 * study_hours + np.random.normal(0, 9, n_students)
exam_score = exam_score.clip(38, 98).round(1)

df = pd.DataFrame({
    'study_hours': study_hours,
    'exam_score': exam_score,
    'gender': np.random.choice(['Male', 'Female'], n_students)
})

# ── OR load your own collected data (uncomment when ready) ───────────
# df = pd.read_csv('/content/your_student_data.csv')
# # Make sure it has columns: study_hours, exam_score (and optionally gender)

print("Dataset preview:")
display(df.head(8))

print(f"Total students: {len(df)}")

# ── 3. Descriptive Statistics ────────────────────────────────────────
stats_table = df[['study_hours', 'exam_score']].describe()

stats_table.loc['median'] = df[['study_hours', 'exam_score']].median()
stats_table.loc['mode']   = df[['study_hours', 'exam_score']].mode().iloc[0]
stats_table.loc['var']    = df[['study_hours', 'exam_score']].var()
stats_table.loc['IQR']    = df[['study_hours', 'exam_score']].quantile(0.75) - df[['study_hours', 'exam_score']].quantile(0.25)
stats_table.loc['CV_%']   = (df[['study_hours', 'exam_score']].std() / df[['study_hours', 'exam_score']].mean() * 100).round(1)

print("\nDescriptive Statistics:")
display(stats_table.round(2))

# ── 4. Skewness & Kurtosis ───────────────────────────────────────────
print("\nShape of distributions:")
print(f"Study hours → Skew = {df['study_hours'].skew():.3f} | Kurtosis = {df['study_hours'].kurtosis():.3f}")
print(f"Exam scores  → Skew = {df['exam_score'].skew():.3f}  | Kurtosis = {df['exam_score'].kurtosis():.3f}")

# ── 5. Visualizations ────────────────────────────────────────────────
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
sns.histplot(df['study_hours'], kde=True, color='dodgerblue', bins=12)
plt.title('Weekly Study Hours')

plt.subplot(2, 2, 2)
sns.histplot(df['exam_score'], kde=True, color='coral', bins=12)
plt.title('Exam Scores (%)')

plt.subplot(2, 2, 3)
sns.boxplot(y=df['exam_score'], color='lightgreen')
plt.title('Boxplot – Exam Scores')

plt.subplot(2, 2, 4)
sns.scatterplot(x='study_hours', y='exam_score', hue='gender', size='exam_score',
                data=df, alpha=0.75, palette='viridis')
plt.title('Study Hours vs Exam Score')

plt.tight_layout()
plt.show()

# ── 6. Correlation ───────────────────────────────────────────────────
r = df['study_hours'].corr(df['exam_score'])
print(f"\nPearson correlation (r) = {r:.3f}")

# ── 7. Simple Linear Regression ─────────────────────────────────────
X = df['study_hours']
y = df['exam_score']

X = sm.add_constant(X)  # adds intercept
model = sm.OLS(y, X).fit()

print("\nRegression results:")
print(f"Equation: Score = {model.params[0]:.2f} + {model.params[1]:.2f} × hours")
print(f"R² = {model.rsquared:.3f}")

sns.regplot(x='study_hours', y='exam_score', data=df,
            line_kws={'color':'darkred'}, scatter_kws={'alpha':0.6})
plt.title('Regression Line + Data Points')
plt.show()

# ── 8. Hypothesis Test (example: <10h vs ≥10h) ──────────────────────
low_group  = df[df['study_hours'] < 10]['exam_score']
high_group = df[df['study_hours'] >= 10]['exam_score']

print(f"\nLow group  (<10h)  → n={len(low_group)}, mean={low_group.mean():.1f}%")
print(f"High group (≥10h) → n={len(high_group)}, mean={high_group.mean():.1f}%")

t_stat, p_val = stats.ttest_ind(low_group, high_group, equal_var=False)

print(f"\nWelch's t-test:")
print(f"  t = {t_stat:.3f}    p = {p_val:.5f}")

if p_val < 0.05:
    print("→ Significant difference (p < 0.05)")
else:
    print("→ No significant difference detected")

# Comparison plot
sns.boxplot(data=[low_group, high_group], palette='Set2')
plt.xticks([0, 1], ['< 10 hours', '≥ 10 hours'])
plt.ylabel('Exam Score (%)')
plt.title('Scores: Low vs High Study Time')
plt.show()

print("\n" + "═"*65)
print("Mini-project ready! Use tables, plots & numbers in your report.")
print("Replace sample data → upload real CSV when you collect responses.")
print("═"*65)
```
